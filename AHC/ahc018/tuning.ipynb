```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
C = [1, 2, 4, 8, 16, 32, 64, 128]
P = [x for x in range(10, 1001, 10)]
```

```python
s: str
with open("./out/0000.txt", "r") as f:
    s = f.read()

l = list(s.split("\n"))
```

```python
x = [list(map(int, m.split())) for m in l]
x = np.array(x)
```

```python
plt.subplot()

for a in range(8):
    plt.plot(P, x[a], label=f"C={C[a]}")

plt.yscale('log')
plt.xscale('log')
plt.legend()

plt.show()
```

```python
for i in range(8):
    print(f"i={i}, C={C[i]}, argmin={x[i].argmin()}, P={P[x[i].argmin()]}")
```

```python
df = pd.DataFrame(
    [[0 for _ in range(20)]  for c in range(len(C))],
    index=[c for c in C],
    columns=[_ for _ in range(20)]
)

for i in range(20):
    s: str
    if i < 10:
        with open(f"./out/000{i}.txt", "r") as f:
            s = f.read()
    else:
        with open(f"./out/00{i}.txt", "r") as f:
            s = f.read()   

    l = list(s.split("\n"))

    x = [list(map(int, m.split())) for m in l]
    x = np.array(x)

    for j in range(8):
        c = C[j]
        df[i][c] = P[x[j].argmin()]

df
```

```python
for c in C:
    print(c, ":", df.loc[c, 0:10].mean(), df.loc[c, 0:20].mean())
```
