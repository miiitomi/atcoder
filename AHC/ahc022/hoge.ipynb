```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("./test_h.txt",sep=" ")
df
```

```python
df["high_value"] = 1
h = 10
for i in range(10000):
    df.loc[i, "high_value"] = h
    if (i+1)%100 == 0:
        h += 10
df
```

```python
sns.barplot(data=df, x="high_value", y="total_score")
```

```python
sns.barplot(data=df.loc[df["high_value"]<=300], x="high_value", y="total_score")
```

```python
tmp = pd.read_csv("./test_s10.txt",sep=" ")
h = 50
tmp["high_value"] = h
for i in range(len(tmp)):
    tmp.loc[i, "high_value"] = h
    if (i+1)%100 == 0:
        h += 50
sns.barplot(data=tmp,x="high_value",y="total_score")
plt.show()
```

```python
tmp = pd.read_csv("./retest_k.txt",sep=" ")
k = 1
tmp["K"] = 0.1
for i in range(len(tmp)):
    tmp.loc[i, "K"] = k/10.0
    if (i+1)%100 == 0:
        k += 1
sns.barplot(data=tmp,x="K",y="total_score")
plt.show()
```

```python
sns.barplot(data=tmp.loc[tmp["K"] >= 3.0],x="K",y="total_score")
```
