```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
from matplotlib import logging; logging.getLogger('matplotlib.font_manager').disabled = True
```

```python
Ns = [60, 80, 100]
Ls = [10, 30, 50]

df = pd.read_csv("./res.txt",sep=" ")
df
```

```python
fig, axes = plt.subplots(3, 3, figsize=(40,40))
for i in range(3):
    for j in range(3):
        N = Ns[i]
        L = Ls[j]
        sns.barplot(data=df.loc[(df["N"]==N)&(df["L"]==L)], x="S", y="total_score", ax=axes[i][j])
        axes[i][j].set_title(f"N={N}, L={L}")

plt.show()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(40,40))
for i in range(3):
    for j in range(3):
        N = Ns[i]
        L = Ls[j]
        sns.barplot(data=df.loc[(df["N"]==N)&(df["L"]==L)&(200 < df["S"])], x="S", y="num_wrong", ax=axes[i][j])
        axes[i][j].set_title(f"N={N}, L={L}")

plt.show()
```
